## Data Visualization

- It is the visual representation of data to discover trends and patterns
- Helps in storytelling and decision making

### Types of Data Visualization

- Categorical: Bar Chart, Pie Chart, etc.
- Numerical: Box plot, Histogram, line plot etc. Line plot is also used with time series data.
- Relationship: Scatter plot, Line plot, Heatmap, Pair plot, etc. looking at the relationship between two variables; if it's more than 2 variables, make use of heatmap

### Libraries

- Matplotlib (it's always good to understand matplotlib first). Seaborn is dependent on matplotlib
- Seaborn

```python
import pandas as pd
import numpy as np

#the pyplot module is all you need in the matplotlib library
import matplotlib.pyplot as plt
```

There are two entities you need to understand in matplotlib; figure and axes

- Figure: It's the entire canvas that holds all subplots e.g. the equivalent of this in excel is the workbook while the equivalent of axes in excel is the worksheet.
- Axes: It's a single plot with x and y axis
- In a figure, you can have multiple plots(they are known as subplots).
- A figure can hold plots like 1, 2, 3, 4, etc.

```python
#Creating a figure with a single axes

x =[1, 2, 3]

y = [4, 5, 6]  #if you want to create a figure with an axes, we have to initialize that first,and to do that;

fig, ax = plt.subplots()    #it is not a must to name it fig.and ax, you can change the variable names (fig, ax) to whatever you like. This subplot method returns two values; one value should be the figure while the other should be for the axes
ax.plot(x, y)

#to display the plot, always remember to call the show function

plt.show()
```

```python
#Always ensure you have titles to your plot; correctly name both your x and y axes
```

```python
#to set title 

x =[1, 2, 3]

y = [4, 5, 6]  #if you want to create a figure with an axes, we have to initialize that first,and to do that;

fig, ax = plt.subplots()    #it is not a must to name it fig.and ax, you can change the variable names (fig, ax) to whatever you like.
ax.plot(x, y)
ax.set_title("Basic Plot")  #add whatever you want to name your plot as an arguement in the set_title method. (we use the set_title method for setting the title)
ax.set_xlabel("X-axis")    #to set the name of x-axis
ax.set_ylabel("Y-axis")    #to set the name of the y-axis

#to display the plot, always remember to call the show function

plt.show()
```

```python
#create a figure with multiple subplots


fig, axes = plt.subplots(1, 2)  #two plots side by side
axes[0].plot(x, y)   #using zero-indexing to get index 0.
axes[1].plot(x,y)

plt.show()

#2 plots side by side, that indicates one row, two columns, while a plot with 2 lines on top of each other is two rows, one columns
```

```python
#Have a picture of how you want your plot(s) to be arranged
```

```python
fig, axes = plt.subplots(1, 2)  #two plots side by side
axes[0].plot(x, y)   #using zero-indexing to get index 0 or access the plot by indexing
axes[0].set_title("Left Plot")
axes[1].plot(x,y)
axes[1].set_title("Right Plot")

plt.show()
```

```python
#to create 4 plots i.e. two rows, two columns

fig, axes = plt.subplots(2, 2)

#to access the first plot, which will be at row 0, column 0 and so on

axes[0, 0].plot(x,y)
axes[0, 1].plot(x,y)
axes[1, 0].plot(x,y)
axes[1, 1].plot(x,y)

plt.show()
```

The dataset showed different netflix shows that have been released over the years
- What are the number of netflix titles released per year?

```python
#grouping by year first. You have to first prepare your data, i.e. the data you want to pass into the plot either by grouping, pivoting, aggregating.

title_per_year = data.groupby("release_year")["title"].count().reset_index()  #starting the grouping with the release year as it is the x-axes, count is one of the aggregating functions

title_per_year
```

```python
#to plot it on the graph, since x and y axes contains numerical data, you can use a line graph; the release year is x-axis, while the y-axis is title
#note: it's not everytime u should set up the fig and axes; and call the sub-plot especially if you are working with just one plot. Always have an imagination of what you want the plot to look like before plotting it in python
#there are different ways to use matplotlib, it all depends on the one faster for you, and more accurate for the type of problem being worked on.
#there are dfferent ways to plot the graph, it depends on what it's comfortable for the programmer.

plt.plot(title_per_year["release_year"], title_per_year["title"])
plt.title("Number of Netflix Titles Released Per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()

#how to interpret the chart below: there is a spike increase among titles released from year 2000 (based on the graph depicted below), netflix experienced a decline in 2020 post-covid
```

```python
#line plot (to beautify the plot, you can change the color of the line, etc. check the matplotlib documentation for more plot editing styles and commands 
#marker is to make the intersection points visible
plt.plot(title_per_year["release_year"], title_per_year["title"], color="Green", marker="o")
plt.title("Number of Netflix Titles Released Per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
data["type"].value_counts()
```

```python
#the data in type is a categorical data, so we can't use a line plot. We can use bar or pie plot/chart. Categorical data rep. variables that fit into groups or labels, rather than numerical measurements.
#piecharts are majorly used for data with 2 or 3 categories, but when it's more than that, bar plot is mostly applicable as the angles in the pie chart wouldn't be readable. Pie chart sums to 360 degrees
#to create a pie chart:

type_counts = data["type"].value_counts() #this indicates the count of each category because that's what we create the segment for the chart 
```

```python
type_counts.index
```

```python
type_counts
```

```python
#Always prepare the data you want to feed into the plot, bar or pie chart during the plotting function/command

plt.pie(type_counts, labels=type_counts.index, autopct='%1.2f%%', colors=["forestgreen", "lightgreen"])
plt.title("Distribution of Movies and TV Shows")
plt.show()  #the autopct is used to find the percentage; the only thing that changes is 2f based on the decimal points you want in your chart
```

```python
#changing the decimal point to 3f

plt.pie(type_counts, labels=type_counts.index, autopct='%1.3f%%', colors=["navy", "skyblue"])
plt.title("Distribution of Movies and TV Shows")
plt.show()

#Try not to use different or varieties of colors for a chart e.g. red and green
#you can change the background of a plot, text color, etc. There's so many editing that can be done on a plot.
```

Bivariate Analysis


- Examining the relationship between two variables

```python
data["type"].value_counts()
```

```python
#movies and tv shows released per year

release_year_by_type = data.groupby(["release_year", "type"]).size()

release_year_by_type

#the size is counting for a particular set of period e.g. in 2020, it counts for both movie and tv shows
```

```python
#movies and tv shows released per year
#we use the unstack method to turn the series into a dataframe; fill_value to change the NaN with 0

release_year_by_type = data.groupby(["release_year", "type"]).size().unstack(fill_value=0).reset_index()

release_year_by_type
```

```python
#Always specify your x-axis before the y-axis during plotting
```

```python
plt.plot(release_year_by_type["release_year"], release_year_by_type["Movie"], label="Movies")
plt.plot(release_year_by_type["release_year"], release_year_by_type["TV Show"], label="TV Shows")
plt.legend()

plt.show()
```
